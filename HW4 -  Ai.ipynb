```python
# cite: https://github.com/jbhender/Stats507_F21/blob/main/ps/ps1_solution.ipynb

def ci_prop(
    x,
    level=0.95,
    str_fmt="{mean:.2f} [{level:.0f}%: ({lwr:.2f}, {upr:.2f})]",
    method="Normal"
):
    # check input type
    try:
        x = np.asarray(x)  # or np.array() as instructed.
    except TypeError:
        print("Could not convert x to type ndarray.")

    # check that x is bool or 0/1
    if x.dtype is np.dtype('bool'):
        pass
    elif not np.logical_or(x == 0, x == 1).all():
        raise TypeError("x should be dtype('bool') or all 0's and 1's.")

    # check method
    assert method in ["Normal", "CP", "Jeffrey", "AC"]

    # determine the length
    n = x.size

    # compute estimate
    if method == 'AC':
        z = norm.ppf(1 - (1 - level) / 2)
        n = (n + z ** 2)
        est = (np.sum(x) + z ** 2 / 2) / n
    else:
        est = np.mean(x)
    # compute bounds for Normal and AC methods
    if method in ['Normal', 'AC']:
        se = np.sqrt(est * (1 - est) / n)
        z = norm.ppf(1 - (1 - level) / 2)
        lwr, upr = est - z * se, est + z * se
    # compute bounds for CP method
    if method == 'CP':
        alpha = 1 - level
        s = np.sum(x)
        lwr = beta.ppf(alpha / 2, s, n - s + 1)
        upr = beta.ppf(1 - alpha / 2, s + 1, n - s)

    # compute bounds for Jeffrey method
    if method == 'Jeffrey':
        alpha = 1 - level
        s = np.sum(x)
        lwr = beta.ppf(alpha / 2, s + 0.5, n - s + 0.5)
        upr = beta.ppf(1 - alpha / 2, s + 0.5, n - s + 0.5)

    # prepare return values
    out = {"mean": est, "level": 100 * level, "lwr": lwr, "upr": upr}
    if str_fmt is None:
        return(out)
    else:
        return(str_fmt.format_map(out))
```

```python
methods = ["Normal", "CP", "Jeffery", "AC"]
nrange = np.arange(10, 20)
prange = np.arange(2, 21)*0.5/20
levels = [0.8, 0.9, 0.95] #choose level = 0,95 for this problem.
nrange, prange = np.meshgrid(nrange, prange)
```

```python
result = []
rng = np.random.default_rng(10*5//2021)

for method in methods:
    result = {}
    estimate_level = np.zeros(nrange.shape)
    width = np.zeros(nrange.shape)
    for i in range(estimate_level.shape[0]):
        for j in range(estimate_level.shape[1]):
            n = nrange[i, j]
            p = prange[i, j]
            # take level = 0.95
            z = norm.ppf((0.95+1)/2)
            n_min = int(np.ceil(z/0.005*np.sqrt(p*(1-p)))**2)
            n_mc = int(n_min +10)
            samples = rng.binomial(n, p, size = n_mc*n).reshape((n_mc, n))
            d = ci_prop(samples.astype('bool'), level = 0.95)
            mean = np.mean(d)
            # mean = d['mean']
            theta = np.sqrt(np.mean(d)*(1-np.mean(d)/n)
            # theta = np.sqrt(d['mean']*(1-d[,mean])/n)
            theta1 = theta!=0
            estimate = np.zeros(n_mc)
            estimate[theta1] = norm.cdf((d['upr'][theta1] - mean[theta1])/
                                        theta[theta1])- norm.cdf((d['lwr'][theta1] - mean[theta1])/theta[theta1])
            estimate_level[i, j] = np.mean(estimate)
            width[i, j] = np.mean(d['upr']-d['lwr'])
    result[method] = {'estimate_level':estimate_level, 'width': width}
result.append(result)
```

```python
fig, axes = plt.subplot(len(methods)//2,2,figsize = (15,15))
axes = axes.reval()
for x, method in enumerate(methods):
    cs = axes[x].counter(nrange, prange, result[x][method]['estimate_level'])
    axes[x].clabel(cs, inline = True)
    axes[x].set_xlabel('n')
    axes[x].set_ylable('p')
    axes[x].setTtitle('CI'% method)
plt.show()
```
