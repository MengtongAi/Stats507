```python


def func0b(sample_list, a, b):
    """
    Use "defaultdict" code to group tuples with the same first element,         
    and then take the maximum value of each group according to the third element.

    Parameters
    ----------
    sample_list : list
        the input of the function
    a : int
        the first paremeter
    b : int
        the second paremeter

    Returns
    -------
    res : list
        
    """
    d = defaultdict(list)
    for i in sample_list:
        d[i[a]].append(i)
    res = [max(val, key=lambda x: x[b]) for val in d.values()]
    return res

func0b([(1, 3, 5), (0, 1, 2), (1, 9, 8)], 0, 2)

# cite: http://nfail.com/q/pTXVlU
```
